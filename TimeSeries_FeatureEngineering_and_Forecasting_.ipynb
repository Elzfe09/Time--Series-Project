## rolling statistics (moving average)
- use sliding time window to calculate mean or other agg value over period defined by that window

kegunaan:
1. smoothing data (reduce noice)
2. detect trend (long term and short term)
3. volatility analysis
4. forecasting

```python
df['cnt_roll7_mean'] = df['cnt'].shift(1).rolling(window =7).mean()
df['cnt_roll7_std'] = df['cnt'].shift(1).rolling(window=7).std()
df
#mean and std of 7 previous value for given attribute
```

## Differencing

Lagging: pakai data lampau sebagai variabel tambahan.

Rolling: ambil rata-rata/sum data lampau dalam jangka waktu tertentu.

Differencing: hitung perubahan antar periode supaya tren hilang.

```python
df['cnt_diff1'] = df['cnt'] - df['cnt'].shift(1)
df['cnt_diff7'] = df['cnt'] - df['cnt'].shift(7)
df
#calculating the diff between present value and its value n times back
#useful: reveal how values change over time
```

# Chile Analysis

```python
data_chile = data[data['Country'] =='Chile']
#sort by date
data_chile.sort_values(by= 'Date', inplace = True)

data_chile.set_index('Date', inplace=  True)
data_chile.drop(columns = ['AirportName','City','State','Country'], inplace = True)
data_chile
```

```python
plt.figure(figsize = (20,10))
plt.plot(data_chile['PercentOfBaseline'])
plt.title('Plot of POB vs Time Period for Chile')
plt.tight_layout()
plt.show()
```

## Dickey-Fuller Test:
- one of statistical test for checking stationarity
- h0: TS is not stationary
- h1: TS is stationery
if p-value less than critical value, rej h0, then TS is stationery

 h1 is something that we want to prove / claim

```python
from statsmodels.tsa.stattools import adfuller
print('Results of Dickey-Fuller Test:')
dftest = adfuller(data_chile['PercentOfBaseline'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)

#non rej h0,TS not stationery
#need to make it stationery bcs all time series models are stationery
#Sifat statistik dari data (mean, variance, autocorrelation) tidak berubah terhadap waktu.
```

# KPSS test

Juga dipakai untuk mengecek stationarity, tapi logika hipotesisnya kebalikan ADF.

Hipotesisnya:

H0: Data stationer.

H1: Data tidak stationer.

Jadi kalau p-value < 0.05 → kita tolak H0, artinya data tidak stationer.

#dipakai dua duanya (DF dan KPSS agar lebih yakin)


```python
#define function for kpss test
from statsmodels.tsa.stattools import kpss
#define KPSS
print ('Results of KPSS Test:')
kpsstest = kpss(data_chile['PercentOfBaseline'], regression='c')
kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
for key,value in kpsstest[3].items():
    kpss_output['Critical Value (%s)'%key] = value
print (kpss_output)

#p-value less than 0.05 rej h0, data tdk stationer
#dicek oleh kedua tes sdh yakin
```

# Solusi:
- differencing: menghitung perbedaan 2 angka dr sblmny dn skrng

fungsi: menghilangkan tren dan seasonality

```python
data_chile['diff'] = data_chile['PercentOfBaseline'] - data_chile['PercentOfBaseline'].shift(1)
plt.figure(figsize=(20,10))
plt.plot(data_chile['diff'])
plt.title('Plot for lagged POB vs Time of Chile')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
#memecah time series jd 3 komponen
decomposition = seasonal_decompose(x=data_chile['PercentOfBaseline'].dropna(),model='multiplicative', period=7)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
plt.figure(figsize=(10,10))
plt.subplot(411)
plt.plot(data_chile['PercentOfBaseline'], label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(data_chile['diff'].dropna(), zero = False)
plt.xlim(0,20)
plt.xticks(np.arange(0,20,1))
plt.show()
#hitung autocorrelation: sbrp hub data hari ini dn sblmny
```

```python
plot_pacf(data_chile['diff'].dropna(),
          zero = False, lags = 40,method = 'ols',
          alpha = 0.05)
plt.xticks(np.arange(0,40,2))
plt.show()
```

```python

from statsmodels.tsa.arima.model import ARIMA  # Pakai ARIMA terbaru

# Buat DataFrame
df = pd.DataFrame(data_chile['diff'])
df.dropna(inplace=True)

# Split data
train = df.iloc[:212]
test = df.iloc[212:]

# Build Model ARMA (pakai ARIMA dengan d=0)
model = ARIMA(train, order=(6,0,0))
fitted = model.fit()
print(fitted.summary())

# Forecast
n_forecast = len(test)  # forecast sepanjang test
forecast_result = fitted.get_forecast(steps=n_forecast)
fc = forecast_result.predicted_mean
conf = forecast_result.conf_int()

# Buat index forecast sesuai test
fc.index = test.index
conf.index = test.index

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='Training', color='blue')
plt.plot(test, label='Actual', color='red')
plt.plot(fc, label='Forecast', color='green')  # Forecast hijau
plt.fill_between(test.index, conf.iloc[:, 0], conf.iloc[:, 1],
                 color='green', alpha=0.2)
plt.title('Forecast vs Actuals')
plt.legend(loc='best', fontsize=8)
plt.show()
```

# Time Series on Airline Passenger

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller, kpss #check stationery data
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose #extract into trend, season, cyclic, random var
from statsmodels.tsa.arima.model import ARIMA #model use
from sklearn.metrics import mean_squared_error # Import mean_squared_error
from math import sqrt
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, index_col = 'Month', parse_dates = True)
# data jlh pelanggan dr wkt ke wkt

plt.figure(figsize=(12,10))
plt.plot(data['Passengers'])
plt.title('Air Passenger Time Series')
plt.xlabel('Time')
plt.ylabel('No of Passengers')

#it give trend and seasonality
```

```python
#plot autocorrelation and partial autocorrelation to check stationerity of data
plot_acf(data['Passengers'].dropna())
plt.show()

#indicate data is non stationer
```

```python
plot_pacf(data['Passengers'])
```

```python
#differencing method
data['Passengers_Diff'] = data['Passengers'].diff()
data.dropna(inplace= True)

plt.figure(figsize=(10,12))
plt.plot(data['Passengers_Diff'])
```

```python
#check whether alr stationer
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(' %s: %.3f' % (key, value))
print("ADF Test Results:")
adf_test(data['Passengers'])
print('\nADF Test Result after Differencing:')
adf_test(data['Passengers_Diff'])
#still more than 0.05 p-value, use second diff
```

## Second differencing

```python
data['Passengers_diffdiff'] = data['Passengers'].diff().diff()
data.dropna(inplace= True)

plot_acf(data['Passengers_diffdiff'])
```

```python
plot_pacf(data['Passengers_diffdiff'])
```

## ARIMA Model

```python
train_data = data['Passengers'][:-36]
test_data = data['Passengers'][-36:]

model = ARIMA(train_data, order=(13,2,2),freq='MS')
#order adalah parameter arima yg berurutan dr p,d,q
model_fit = model.fit()
model_fit.summary()
```

```python
predictions = model_fit.predict(start = len(train_data),
                                end = len(data)-1)
rmse = sqrt(mean_squared_error(test_data, predictions))
print(f'\nRMSE:{rmse}')
```

```python
#predictions vs actual
plt.figure(figsize=(12,6))
plt.plot(test_data, label = 'Actual')
plt.plot(predictions, color = 'red', label = 'Predicted')
plt.title('Air Passengers Forecasting')
```

# Result
model understands pattern, but this is seasonal data it require few more tweaks to match pattern

comparing stationery vs non-stationery data , leaving it non-stationery

```python
model = ARIMA(train_data, order=(13,0,2), freq='MS')
model_fit = model.fit()
model_fit.summary()
predictions= model_fit.predict(start=len(train_data),
                               end = len(data)-1)
rmse = sqrt(mean_squared_error(test_data, predictions))
print(f'\nRMSE: {rmse}')

# Visualize predictions vs actual
plt.figure(figsize=(12, 6))
plt.plot(test_data, label='Actual')
plt.plot(predictions, color='red', label='Predicted')
plt.title('Air Passengers Forecasting')
plt.xlabel('Time')
plt.ylabel('Passengers')
plt.legend()
plt.show()
```

q =  → Orde MA (Moving Average)
